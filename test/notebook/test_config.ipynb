## Test 3: Supported Package Formats

```python
import pandas as pd

# Create a DataFrame of supported formats
formats_data = []
for key, value in config.SUPPORTED_FORMATS.items():
    formats_data.append({
        'Format': key,
        'Extension': value['extension'],
        'Description': value['description'],
        'Icon': value['icon'],
        'Managers': ', '.join(value['managers'])
    })

df = pd.DataFrame(formats_data)
print("Supported Package Formats:\n")
print(df.to_string(index=False))

# Assertions
assert 'deb' in config.SUPPORTED_FORMATS, "Should support .deb packages"
assert 'rpm' in config.SUPPORTED_FORMATS, "Should support .rpm packages"
assert 'snap' in config.SUPPORTED_FORMATS, "Should support .snap packages"
assert 'flatpak' in config.SUPPORTED_FORMATS, "Should support .flatpak packages"
print("\nâœ… Package format tests passed")
```

## Test 4: Helper Functions

```python
# Test get_supported_extensions()
extensions = config.get_supported_extensions()
print("Supported Extensions:")
for ext in extensions:
    print(f"  {ext}")

assert '.deb' in extensions, ".deb should be supported"
assert '.rpm' in extensions, ".rpm should be supported"
assert '.snap' in extensions, ".snap should be supported"
assert '.flatpak' in extensions, ".flatpak should be supported"
print("\nâœ… Extension tests passed")
```

```python
# Test is_supported_package()
test_files = [
    ('package.deb', True),
    ('package.rpm', True),
    ('package.snap', True),
    ('package.flatpak', True),
    ('package.txt', False),
    ('package.exe', False),
    ('PACKAGE.DEB', True),  # Case insensitive
]

print("Testing is_supported_package():\n")
for filename, expected in test_files:
    result = config.is_supported_package(filename)
    status = "âœ…" if result == expected else "âŒ"
    print(f"  {status} {filename}: {result} (expected {expected})")
    assert result == expected, f"Failed for {filename}"

print("\nâœ… Package support tests passed")
```
